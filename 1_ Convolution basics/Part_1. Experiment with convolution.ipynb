#### Identity

```python
from scipy.signal import convolve2d
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

data1   = cv2.imread('dog.jpg',cv2.IMREAD_GRAYSCALE)
kernel = np.array([[0,0,0,],
                   [0,1,0],
                   [0,0,0]])

print(data1.shape)
print(kernel.shape)

b=convolve2d(data1,kernel)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(131).imshow(data1,'gray')
plt.subplot(132).imshow(kernel,'gray')
plt.subplot(133).imshow(b,'gray')
```
